ВВП 1950-2010

```
from matplotlib import pyplot as plt
from collections import Counter

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gpd = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gpd, color='green', marker='o', linestyle='solid')

plt.title('Номинальный ВВП')
plt.ylabel('Млрд $')

plt.show
```

```
movies = ["Останься", "Drive", "Порожденный", "Место под соснами", "Отступники"]
oscars = [5, 13, 1, 8, 10]
xs = [i + 0.1 for i, _ in enumerate(movies)]
plt.bar(xs, oscars)

plt.ylabel("Колличество наград")
plt.title("Мои любимые фильмы")

plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()],
        histogram.values(),
        8)

plt.axis([-5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Дециль')
plt.ylabel('Число студентов')
plt.title('Распределение оценок за экзамен № 1')
plt.show()
```

```
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks(years)
plt.ylabel("Больше не такой огромный")

plt.ticklabel_format(useOffset=False)
plt.axis([2012.5, 2014.5, 0, 550])
plt.show()
```

```
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x+y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
"""метод plt можно вызывать много раз
чтобы вызвать несколько графиков на одной и той же диаграмме:
зеленаR сплошная линия"""

plt.plot(xs, variance, 'g-', label='дисперсия')
plt.plot(xs, bias_squared, 'r-.', label='смещение^2')
plt.plot(xs, total_error, 'b:', label='суммарная ошибка')

plt.legend(loc=9)
plt.xlabel("Сложность модели")
plt.title("компромисс между смещением и дисперсией")
plt.show()
```
